```python
import os
import random
import cv2
from PIL import Image

# 1. Load YOUR newly trained brain! (YOLO saved your best weights here)
my_model = YOLO('runs/detect/train/weights/best.pt')

# 2. Pick a random image from the validation folder (an image it hasn't studied)
valid_folder = f"{dataset.location}/valid/images/"
random_pic = random.choice(os.listdir(valid_folder))
image_path = f"{valid_folder}/{random_pic}"

# 3. Tell your model to scan the raw image and draw boxes!
print(f"Testing on image: {random_pic}")
results = my_model.predict(source=image_path, conf=0.25) # conf=0.25 means "be at least 25% confident"

# 4. Extract the image with the YOLO boxes drawn on it and display it
result_array = results[0].plot()
final_image = Image.fromarray(cv2.cvtColor(result_array, cv2.COLOR_BGR2RGB))
final_image
```

```python
# 1. Ask the brain what classes it knows
print("Classes my model knows:", my_model.names)

# 2. Run the scan again, but let it show boxes even if it's only 10% confident
print(f"Testing on image: {random_pic} (Lowering confidence to 10%)")
results = my_model.predict(source=image_path, conf=0.10)

# 3. Show the image
result_array = results[0].plot()
final_image = Image.fromarray(cv2.cvtColor(result_array, cv2.COLOR_BGR2RGB))
final_image
```

```python
import matplotlib.pyplot as plt

# 1. Grab the raw image again
original_img = cv2.imread(image_path)
original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)

# 2. Extract the math coordinates of the boxes YOLO found
# xyxy means [x_min, y_min, x_max, y_max]
boxes = results[0].boxes.xyxy.cpu().numpy()

# 3. Take out the virtual scissors and cut!
if len(boxes) == 0:
    print("YOLO didn't find anything to crop!")
else:
    print(f"✂️ YOLO found {len(boxes)} object(s)! Cutting them out now...\n")
    for i, box in enumerate(boxes):
        # Convert the math coordinates into whole numbers (pixels)
        x1, y1, x2, y2 = map(int, box)

        # Crop the image array using those coordinates
        cropped_piece = original_img[y1:y2, x1:x2]

        # Display the cropped piece
        print(f"Crop #{i+1}:")
        plt.imshow(cropped_piece)
        plt.axis('off') # Hide the messy axes
        plt.show()
```
